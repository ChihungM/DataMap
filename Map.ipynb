```python
import folium
import numpy
 
m = folium.Map(location=[52.5695, -0.2405], zoom_start=12)
 
tooltip = "Click for more info!"

with open("text.txt") as f:
    for line in f:
        c1, c2, address = line.split(",", 2)
        folium.Marker([float(c1), float(c2)],popup=address).add_to(m)

m.save("index.html")
```
